```python
def save_file(path, file_to_save):
    with open(path, "w", encoding="utf-8") as f:
        f.write(file_to_save)
```

```python
def read_file(path):
    with open(path, "r", encoding="utf-8") as f:
        data = json.loads(f.read())
    print(f"length of file = {len(data)}")
    return data
```

```python
def update_verse(updated_verse):
    progression = range(0, 1000, 100)
    count = 0
    for vrs in updated_verse:
        verse = Verse.objects.get(pk=vrs["pk"])
        verse.text = vrs["fields"]["text"]
        verse.lemmas = vrs["fields"]["lemmas"]
        verse.forms_by_lemma = vrs["fields"]["forms_by_lemma"]
        verse.save()

        if count in progression:
            print(count)
        count += 1
```

```python
def update_word(words_updated):
    progression = range(0, 1000, 100)
    count = 0
    for w in words_updated:
        word = Word.objects.get(pk=w["pk"])
        word.neologism = w["fields"]["neologism"]
        word.quotes = w["fields"]["qoutes"]
        word.forms = w["fields"]["forms"]
        word.save()
        if count in progression:
            print(count)
        count += 1
        
```

```python
def update_trans(updated_trans):
    progression = range(0, 1000, 100)
    count = 0
    for w in updated_trans:
        WordTranslation.objects.create(
            pk=w["pk"]),
            word = w["fields"]["word"],
            lang = w["fields"]["lang"],
            trans = w["fields"]["trans"],
            is_preferred = w["fields"]["is_preferred"],
        )
        if count in progression:
            print(count)
        count += 1
        
```
